# Estudos da biblioteca matplotlib.pyplot

#### Anotações e observações no final

````python
# Gráfico de linha

# Importar a biblioteca

import matplotlib.pyplot as plt

# Variáveis guardando listas que definirão os pontos dos eixos

x = [1, 2]
y = [2, 3]

# Legendas 

plt.title("Gráfico de linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# Plotar o gráfico de linha

plt.plot(x, y)
plt.show()
````

````python
# Gráfico de barras 1
# Apenas um par de variáveis X e Y

# Importar a biblioteca

import matplotlib.pyplot as plt

# Variáveis guardando listas que definirão os pontos dos eixos

x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

# Variáveis guardando strings com as legendas do gráfico

titulo = "Gráfico de barras"
eixo_x = "Eixo X"
eixo_y = "Eixo Y"

# Definição das legendas

plt.title(titulo)
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

# Plotar o gráfico de barras

plt.bar(x, y)
plt.show()
````

````python
# Gráfico de barras 2
# Comparando dois eixos X e dois eixos Y

# Importar a biblioteca

import matplotlib.pyplot as plt

# Variáveis guardando listas que definirão os pontos dos eixos

x1 = range(1, 10, 2)
y1 = [2, 3, 7, 1, 0]

x2 = range(2, 11, 2)
y2 = [5, 1, 3, 7, 4]

# Variáveis guardando strings com as legendas do gráfico

titulo = "Gráfico de barras"
eixo_x = "Eixo X"
eixo_y = "Eixo Y"

# Definição das legendas

plt.title(titulo)
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

# Plotar o gráfico de barras

plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend() # adiciona uma caixa explicando as cores 
plt.show()
````

````python
# Gráfico de dispersão (scatterplot)

import matplotlib.pyplot as plt

x = range(1, 10, 2)
y = [8, 3, 2, 0, 5]


titulo = "Gráfico de dispersão"
eixo_x = "Eixo X"
eixo_y = "Eixo Y"


plt.title(titulo)
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

# Plotar o gráfico de dispersão

plt.scatter(x, y) # Plota somente os pontos
plt.plot(x, y)    # Plota também uma linha interligando os pontos
plt.show()

# Se quiser, pode trocar as cores dos pontos ou das linhas.
# Basta adicionar o parâmetro color = 'código da cor' no argumento dos plots.
# Ex.: plt.scatter(x, y, color = 'r') para ter pontos vermelhos
# https://matplotlib.org/stable/tutorials/colors/colors.html
````

# Comentários

- #### Sintaxe python:  Biblioteca.Módulo.Função(Parâmetros)
    Ex.:```matplotlib.pyplot.xlabel()```

#### Tutorial oficial matplotlib: https://matplotlib.org/stable/tutorials/index.html

- Usar função ```range()``` para ajudar no eixo X:
    - A função inclui o valor inicial, porém exclui o valor final.
    - ```range(Valor Inicial, Valor Final + 1, Passo)```
    - Obs.: Se for necessário salvar a variável como uma lista, aplicar o range dentro da função ```list()```.

#### A seguir, alguns exemplos de argumentos que podem ser aplicados ao método plot( ).

- color: cor (ver exemplos abaixo)

- label: rótulo

- linestyle: estilo de linha (ver exemplos abaixo)

- linewidth: largura da linha

- marker: marcador (ver exemplos abaixo)



#### CORES (color)
'b' blue

'g' green

'r' red

'c' cyan

'm' magenta

'y' yellow

'k' black

'w' white



#### Marcadores (marker)

',' pixel marker

'^' triangle_up marker

'<' triangle_left marker

'>' triangle_right marker

'1' tri_down marker

'2' tri_up marker

'3' tri_left marker

'4' tri_right marker

's' square marker

'p' pentagon marker

'*' star marker

'h' hexagon1 marker

'H' hexagon2 marker

'+' plus marker

'x' x marker

'D' diamond marker

'|' vline marker
